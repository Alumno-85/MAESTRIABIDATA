# Ejercicio 1

Para realizar el ejercicio cargamos los datos de *Especies* en STARWARS SWAPI y las librerías principales.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()  # para el estilo de graficos

entidades = ['planets','starships','vehicles','people','species']
entidades_df = {x: pd.read_pickle('www/' + x + '_df.pkl') for x in entidades}

# Datos principales
people_df = entidades_df['people'][["height","mass","birth_year","gender","homeworld"]].dropna()

people_df
```

## Ejercicio 1.1.

## Ejercicio 1.5.

Calcula, a partir de los vectores numpy de altura y peso, vectores numpy que debe definir, el [índice de masa corporal (IMC)](https://es.wikipedia.org/wiki/%C3%8Dndice_de_masa_corporal) de los personajes de star wars contenidos en people_df:

$IMC = \frac{peso}{altura^{2}}$ donde altura está en metros y el peso en kg.

Muestra los datos por pantalla.

```python
# Solución:
people_df_filtered = people_df.dropna(subset=['height', 'mass'])
altura_metros = people_df_filtered['height'] / 100
imc = people_df_filtered['mass'] / (altura_metros ** 2)
people_df_filtered['IMC'] = imc
print(people_df_filtered[['IMC']])
```

## Ejercicio 1.6.

A partir del IMC que has calculado en el ejercicio 1.5. Construye un panel con dos histogramas:

- Un histograma con toda la muestra
- Un histograma seleccionando los valores con un IMC inferior a 100.

```python
# Solución:
fig, axs = plt.subplots(2, 1, figsize=(10, 8))

# Histograma con toda la muestra
axs[0].hist(people_df_filtered['IMC'], bins=20, color='skyblue', edgecolor='black')
axs[0].set_title('HISTOGRAMA DE IMC - TODA LA MUESTRA')
axs[0].set_xlabel('IMC')
axs[0].set_ylabel('No. Personajes')

# Histograma con valores de IMC inferiores a 100
imc_menor_100 = people_df_filtered[people_df_filtered['IMC'] < 100]['IMC']
axs[1].hist(imc_menor_100, bins=20, color='salmon', edgecolor='black')
axs[1].set_title('HISTOGRAMA DEL IMC < 100')
axs[1].set_xlabel('IMC')
axs[1].set_ylabel('No. Personajes')

# Ajustar diseño y mostrar los histogramas
plt.tight_layout()
plt.show()
```

## Ejercicio 1.7.

A partir del vector 1 y 2 que se definen a continuación contesta las siguientes preguntas:

- Calcula el shape, ndim, size del vector1 y vector2
- Explica cual es la diferencia entre vector1 y vector2 a partir de los que hayas observado

```python
vector1 = np.hstack([altura,peso])
vector2 = np.vstack([altura,peso])
```

```python
# Solución:
altura = np.array([160, 170, 165])
peso = np.array([60, 70, 65])
vector1 = np.hstack([altura, peso])
vector2 = np.vstack([altura, peso])
shape_vector1 = vector1.shape
ndim_vector1 = vector1.ndim
size_vector1 = vector1.size
shape_vector2 = vector2.shape
ndim_vector2 = vector2.ndim
size_vector2 = vector2.size
print("Vector 1:")
print("Shape:", shape_vector1)
print("Ndim:", ndim_vector1)
print("Size:", size_vector1)
print("\nVector 2:")
print("Shape:", shape_vector2)
print("Ndim:", ndim_vector2)
print("Size:", size_vector2)

# Conclusión
print("\nConclusión:")
print("- vector1 es un vector unidimensional que contiene todos los elementos de altura seguidos por todos los elementos de peso")
print("- vector2 es una matriz bidimensional donde la primera fila corresponde a los datos de altura y la segunda fila corresponde a los datos de peso")
```

Crea una copia de people_df llamada personajes_df y renombra las columnas con su traducción al castellano.

Muestra los 5 primeros registros del nuevo data frame con `.head()`.

## Ejercicio 1.9.

Haz el mismo cálculo de IMC que has hecho en 1.5. pero directamente sobre el objeto personajes_df.

Ordena el data frame de mayor a menor IMC y muestra el nombre e IMC de los personajes con IMC por encima de 30.

> ¿Sabias que IMC por encima de 30 se considera obeso?


```python
# Solución:
altura_metros = personajes_df['altura'] / 100
peso_kg = personajes_df['peso']
imc = peso_kg / (altura_metros ** 2)
personajes_df['IMC'] = imc
personajes_df = personajes_df.sort_values(by='IMC', ascending=False)
obesos_df = personajes_df[personajes_df['IMC'] > 30]
print("Personajes con IMC por encima de 30 (considerados obesos):")
print(obesos_df[['IMC']])
```

## Ejercicio 1.10.

Inserta un valor faltante en los valores de IMC que sean superiores a 100 y dibuja el histograma del IMC transformado con `.plot.hist()`.

```python
# Solución:
personajes_df.loc[personajes_df['IMC'] > 100, 'IMC'] = np.nan
personajes_df['IMC'].plot.hist(bins=20, color='skyblue', edgecolor='black', figsize=(10, 6))
plt.title('HISTOGRAMA DEL IMC TRANSFORMADO')
plt.xlabel('IMC')
plt.ylabel('No. Personajes')
plt.grid(axis='y', alpha=0.75)
plt.show()
```
